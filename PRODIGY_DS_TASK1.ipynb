```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace the URL with the actual URL of the dataset
url = "/content/API_SP.POP.TOTL_DS2_en_csv_v2_6224560.csv"
data = pd.read_csv(url, skiprows=4)

# Selecting relevant columns
data = data[['Country Name', '2019']]  # You can choose a different year if needed

# Drop rows with missing values
data = data.dropna()

# Sorting data by population in descending order
data = data.sort_values(by='2019', ascending=False)

# Plotting a bar chart
plt.figure(figsize=(12, 8))
plt.bar(data['Country Name'][:10], data['2019'][:10], color='skyblue')
plt.title('Top 10 Countries by Population in 2019')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
